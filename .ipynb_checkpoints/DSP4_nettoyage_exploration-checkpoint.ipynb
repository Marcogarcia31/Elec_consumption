###  Concatenating  Dataframes

We want to concatenate the data of 2015 & 2016 to work on a single dataframe.

#### 2015 'Location' col parsing

To concatenate, we should first parse 'Location' column in 2015 data to match 2016's.

```python
data_2015['Location'].head()
```

```python
import json
### Preprocessing function for json strings of 'Location' column 
def string_processor(x):
    x = x.replace('"', '&')
    x = x.replace("'", '"')
    x = x.replace('&', "'")
    return x


### Parse 2015 'Location' columns into several columns
def location_col_parsing(data):
    
    data = data.copy()
    
    ###Converts 2015 'Location' column values from strings to dictionaries
    data['Location'] = [json.loads(string_processor(location)) for location in data['Location']]
    
    ### Keys of each dictionary in the 'Location' column
    location_keys = ['human_address', 'latitude', 'longitude']
    
    ### Creates new column in the df corresponding to each key of the dictionaries in 'Location' column
    for key in location_keys:
            data[key] = [building[key] for building in data['Location']]
    
    ### Sub-keys of human_address key
    address_keys = ['address', 'city', 'state', 'zip']
    
    ### Creates new column in the df corresponding to each key of the dictionaries in 'Human_addess' column
    for key in address_keys:    
        data[key] = [json.loads(address.replace("'", '"'))[key] for address in data['human_address']]
                
    
    data = data.drop(['Location', 'human_address'], axis = 1)
                                   
    return data
```

```python
data_2015 = location_col_parsing(data_2015)
data_2015.head()
```

Parsing is done.

#### Harmonizing column labels of DataFrames

Columns in both DataFrames represent the same feature but do not have the same label.
We will rename some columns of the 2015 dataframe to match 2016.

```python
### Mapping to match 2016 simpler labels
mapper = {'GHGEmissions(MetricTonsCO2e)' : 'TotalGHGEmissions',
         'GHGEmissionsIntensity(kgCO2e/ft2)' : 'GHGEmissionsIntensity', 
         'address' : 'Address', 
          'city' : 'City', 
          'state': 'State', 
          'zip' : 'ZipCode', 
         'latitude' : 'Latitude', 
          'longitude': 'Longitude'}

data_2015 = data_2015.rename(mapper = mapper, axis = 1)
```

#### Concatenating DataFrames

We will only consider common column labels in both DataFrames for concatenation. We'll assume features present only in one of the two years are not that relevant.

```python
### Function that takes a sequence of DataFrames as input
### looks for common column labels and concatenates row-wise
def concatenation(df_1, df_2):
    
    cols_1 = set(df_1.columns)
    cols_2 = set(df_2.columns)

    ### gets the intersection of column labels
    common_cols = cols_1.intersection(cols_2)

    data = pd.concat([df_1[common_cols], df_2[common_cols]])

    return data
```

```python
data = concatenation(data_2015, data_2016)
print('data has ' + str(data.shape[0]) + ' lines and ' + str(data.shape[1]) + ' columns')
```

Concatenation seems to be done. We will now clean the data.

## Exploration

We will first split the data into train and test splits so that we can later on confirm our exploration insights on an untouched dataset.

### Train/test splitting

We will modify a sample label value on 'PrimaryPropertyType' column to be able to stratify our split based on the occurrences of each label of that column.

```python
### Modifies a building's value so as to able to stratify the split on that feature
mask = data_v5['PrimaryPropertyType'] == 'Restaurant\n'
data_v5['PrimaryPropertyType'][mask] = 'Restaurant'
```

```python
from sklearn.model_selection import train_test_split

def splitting(data):
    
    data = data.copy()

    X = data.drop([energy_target, co2_target], axis = 1)
    
    X_train, X_test, energy_target_train, energy_target_test, co2_target_train, co2_target_test = train_test_split(X, data[energy_target], 
                                                                                data[co2_target], 
                                                                                        test_size=0.2, stratify = X['PrimaryPropertyType'], 
                                                                                    random_state = 42)
    
    return X_train, X_test, energy_target_train, energy_target_test, co2_target_train, co2_target_test
    
```

```python
X_train, X_test, energy_target_train, energy_target_test, co2_target_train, co2_target_test=splitting(data_v5)
```

#### PCA on size features

```python
from  sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

### Plots correlation circle between size building features
def corr_circle(data):
    
    
    data = data.copy()
    
    features = ['SecondLargestPropertyUseTypeGFA', 
            'NumberofBuildings', 'PropertyGFABuilding(s)', 
        'NumberofFloors',
       'PropertyGFAParking', 'LargestPropertyUseTypeGFA']
    
    data = np.log(data[features]+1)
    
    ### instantiate pipeline with simple imputer, scaler & PCA then fit pipeline 
    pipe = Pipeline([('imputer', SimpleImputer(strategy = 'median')),('scaler', StandardScaler()), ('pca', PCA())])
    pipe.fit(data)
    
    ### fitted PCA to address components coordinates on feature space
    pca = pipe.named_steps['pca']
    
    
    fig = plt.figure(figsize = (8,8))
    ax = fig.add_subplot(111)
    
    for i, feature in enumerate(features):
        
        ### for a given feature the arrow coordinates are coefs 
        ### on first & second component by explained variance
        ax.arrow(0,0, pca.components_[0, i], 
            pca.components_[1,i], 
        head_width = 0.02, 
        head_length = 0.01)
        
        ### Include feature label
        ax.text(pca.components_[0,i], 
    pca.components_[1,i], 
    feature, fontsize = 8)
        
        ### Include coordinates on PCs next to feature label
        ax.text(pca.components_[0,i], 
    pca.components_[1,i] - 0.05, 
        '( ' + str(pca.components_[0,i].round(decimals = 2)) + ',' 
    + str(pca.components_[1,i].round(decimals = 2)) + ' )', fontsize = 7)
    
    ### Plot x and y axis on the figure
    zeros = np.zeros(50)
    line = np.linspace(-1, 1, 50)
    ax.plot(zeros,line)
    ax.plot(line,zeros)
    
    an = np.linspace(0, 2 * np.pi, 100)
    ax.plot(np.cos(an), np.sin(an))  # Add a unit circle for scale
    plt.axis('equal')
    ax.set_ylabel(ylabel = 'PC 2' + '  '+ '(' + str("{:.0%}".format(pca.explained_variance_ratio_[1])) + ')', fontsize = 16)
    ax.set_xlabel(xlabel = 'PC 1' + '  '+ '(' + str("{:.0%}".format(pca.explained_variance_ratio_[0])) + ')', fontsize = 16)
    ax.set_title('Correlation circle', fontsize = 24)

corr_circle(X_train)
```

Property GFA Building & Largest Property Use Type GFA seem strongly positively correlated by looking at the correlation circle. Thay have pretty the same direction and length. 

```python
### Consider log-transformed features
np.log(X_train[['LargestPropertyUseTypeGFA', 'PropertyGFABuilding(s)']]+1).corr()
```

### Energy sources

#### Feature engineering on energy source 


As we have information about each enery source consumption for ach building, we will generate two sorts of features by transforming the 'energy source' features(eg. 'Electricity', 'Steam'...) into :
* a categorical feature : main energy source of the building (3 classes)
* three binary features for each energy source: whether the energy source is used in the building or not

#### Source vs size

```python
### Custom module with custom class : creates the features described above
from custom_transformer import Energy_variables_generator

### Displays distribution on targets for each building category based on main energy source
def target_value_per_source(data, targets):
    
    data = data.copy()
    
    generator = Energy_variables_generator()
    data = generator.transform(data)
    fig = plt.figure(figsize = (12,4))
    
    for i, target in enumerate(targets):
        data = pd.concat([data, target], axis = 1)
        g = fig.add_subplot(1,2,i+1)
        g = sns.boxplot(x = 'main_energy_source',
                    y = target.name, data = data)
        g.set_title(target.name, fontsize = 12, pad = 12)

        plt.xticks(rotation = 0,fontsize = 8)
        
        
target_value_per_source(X_train, [log_energy_target_train, log_co2_target_train])
```

We can see on the boxplots that the buildings which use Steam as main source of energy tend to have higher consumption values on both targets. 
* But does the energy source have an impact ? 
* Or is it just the buildings that use Steam are bigger than the others ?

```python
### Linear regression between targets & size for each main_energy_source class
def linear_plot(data, target):
    
    data = data.copy()
    
    fig = plt.figure(figsize = (12,4))
    
    ### Concatenate features with target
    data = pd.concat([data, target], axis = 1)
    
    data['PropertyGFABuilding(s)'] = np.log(data['PropertyGFABuilding(s)']+1)
    ### Generates features
    generator = Energy_variables_generator()
    data = generator.transform(data)
    
    g = sns.lmplot(x = 'PropertyGFABuilding(s)', 
        y = target.name, hue = 'main_energy_source', 
    data = data, legend_out = False)
    fig.suptitle(target.name, fontsize = 16)
    
    
linear_plot(X_train, log_energy_target_train)
```

* The plot here shows that, for a given building size, buildings which use Steam as main source of energy tend to have higher values on energy consumption.
* This suggests that the features created are relevant and might help estimators when fitting a model.

```python
linear_plot(X_train, log_co2_target_train)
```

The insight is even clearer for co2 emissions. For a given building size, buildings which use Steam as main energy source tend to have significantly superior co2 emissions than electricity-driven buildings.
